```python
from pylab import *
from matplotlib import rc
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
%matplotlib notebook
clf()
xoff=0.22
yoff=0.2
width=0.8-xoff
height=0.95-yoff


fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [1*fig_width,2.5*fig_height]
params = {'backend': 'pdf',
          'axes.labelsize': 9,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          #'text.usetex': False,
          'figure.figsize': fig_size,
          'figure.subplot.left': xoff,
          'figure.subplot.bottom': yoff,
          'figure.subplot.right': xoff+width,
          'figure.subplot.top': yoff+height}

rcParams.update(params)
#rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})

rcParams["font.family"] = "Tex Gyre Termes"
rcParams["font.weight"] = "normal"
rcParams["axes.labelweight"] = "normal"

figure(1)

gs1=gridspec.GridSpec(3,1)
gs1.update(hspace=0.0,wspace=0.0)

osc1i=np.loadtxt('osc1_intensity_new.txt',delimiter='\t')
osc2i=np.loadtxt('osc2_intensity_new.txt',delimiter='\t')
osc1p=np.loadtxt('osc1_phase_new.txt',delimiter='\t')
osc2p=np.loadtxt('osc2_phase_new.txt',delimiter='\t')
osc1r=np.loadtxt('osc1_refl_new.txt',delimiter='\t')
osc1cr=np.loadtxt('osc1_coupled_refl_new.txt',delimiter='\t')

subplot(gs1[0,0])

plot([e[0] for e in osc1i],[e[1] for e in osc1i],color='b',lw=1)
plot([e[0] for e in osc2i],[e[1] for e in osc2i],color='r',lw=1)
l1,=plot([2.91677,2.91677],[1,5e3],color='k',lw=0.75)
l1.set_dashes([3,3])
l1,=plot([3.44392,3.44392],[1,5e3],color='k',lw=0.75)
l1.set_dashes([3,3])
gca().set_yscale('log')

setp(gca(),xlim=[2.0,4.0],ylim=[1,5e3],xticklabels=[])
ylabel('Intensity $|A(\omega)|^2$')
gca().tick_params(axis='x',which='both',direction="in")
gca().yaxis.set_label_coords(-0.19,0.5)


subplot(gs1[1,0])

plot([e[0] for e in osc1p],[e[1] for e in osc1p],color='b',lw=1)
plot([e[0] for e in osc2p],[e[1] for e in osc2p],color='r',lw=1)
plot([2,4],[0,0],color='k',lw=0.75)
l1,=plot([2,4],[3.1415/2,3.1415/2],color='k',lw=0.75)
l1.set_dashes([2,2])
l2,=plot([2,4],[-3.1415/2,-3.1415/2],color='k',lw=0.75)
l2.set_dashes([2,2])
l1,=plot([2.91677,2.91677],[-3.2,3.2],color='k',lw=0.75)
l1.set_dashes([3,3])
l1,=plot([3.44392,3.44392],[-3.2,3.2],color='k',lw=0.75)
l1.set_dashes([3,3])


setp(gca(),xlim=[2.0,4.0],ylim=[-2,2],xticklabels=[],yticks=arange(-3.1415/2,3.1415,3.1415/2),yticklabels=[r'$-\pi/2$',r'0',r'$\pi/2$'])
ylabel('Phase $\Phi(\omega)$ [rad]')
gca().tick_params(axis='x',which='both',direction="in")
gca().yaxis.set_label_coords(-0.19,0.5)

subplot(gs1[2,0])

plot([e[0] for e in osc1r],[e[1] for e in osc1r],color='b',lw=1)
plot([e[0] for e in osc1cr],[e[1] for e in osc1cr],color='g',lw=1)
l1,=plot([2.91677,2.91677],[0,1],color='k',lw=0.75)
l1.set_dashes([3,3])
l1,=plot([3.44392,3.44392],[0,1],color='k',lw=0.75)
l1.set_dashes([3,3])

setp(gca(),xlim=[2.0,4.0],ylim=[0,0.9],xticks=arange(2,4.5,0.5))
ylabel('Reflectance')
xlabel('Angular freq. $\omega$ [$10^{15}$ Hz]')
gca().yaxis.set_label_coords(-0.19,0.5)

show()
savefig('oscillators_new.svg',bbox_inches='tight')
```
